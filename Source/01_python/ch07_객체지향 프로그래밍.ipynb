```python
â€» ì¸ìŠ¤í„´ìŠ¤ ë©”ì†Œë“œ vs í´ë˜ìŠ¤ ë©”ì†Œë“œ vs static ë©”ì†Œë“œ
- ì¸ìŠ¤í„´ìŠ¤ë©”ì†Œë“œ : ì²« ë²ˆì§¸ ì¸ìì— selfë¥¼ ê¼­ ëª…ì‹œ, ê°ì²´ë¥¼ í†µí•´ì„œë§Œ í˜¸ì¶œí•  ìˆ˜ ìˆëŠ” ë©”ì†Œë“œ
- í´ë˜ìŠ¤ë©”ì†Œë“œ : ë©”ì†Œë“œìœ„ì— @classmethodë¥¼ ëª…ì‹œ. ì²« ë²ˆì§¸ ì¸ìì— clz(cls)ë¥¼ ëª…ì‹œ
- staticë©”ì†Œë“œ : ë©”ì†Œë“œìœ„ì— @staticmethodë¥¼ ëª…ì‹œ, ë””í´íŠ¸ ì¸ì ì—†ìŒ.
```

```python
class Person:
    academy = 'MBC'   # í´ë˜ìŠ¤ ë³€ìˆ˜
    def __init__(self, name="í™ê¸¸ë™", gender="ë‚¨ì"):
        self.name = name
        self.gender = gender
    def print_info(self):
        print("ì¼ë°˜(ì¸ìŠ¤í„´ìŠ¤)ë©”ì†Œë“œ - {}ì˜ {}ë‹˜ì€ {}ì…ë‹ˆë‹¤".format(self.academy, self.name, self.gender))
    
    def change_info(self, name="ìµëª…", gender="ì—¬ì"):
        self.name = name
        self.gender = gender
        print("ì¼ë°˜(ì¸ìŠ¤í„´ìŠ¤)ë©”ì†Œë“œë¥¼ í†µí•´ {}ë¡œ ì´ë¦„ë³€ê²½. {}ë¡œ ì„±ë³„ ë³€ê²½".format(self.name, self.gender))
    @classmethod
    def do_(clz):# # clz => Person ì˜ë¯¸. í´ë˜ìŠ¤ ë³€ìˆ˜ë¥¼ ì´ìš©í•œ ë¡œì§
        clz.academy = 'classAcademy'
        print("í´ë˜ìŠ¤ ë©”ì†Œë“œì—ì„œ {}ë¡œ í´ë˜ìŠ¤ ë³€ìˆ˜ ë³€ê²½".format(clz.academy))
    
    @classmethod
    def done_(clz, academy):# # clz => Person ì˜ë¯¸. í´ë˜ìŠ¤ ë³€ìˆ˜ë¥¼ ì´ìš©í•œ ë¡œì§
        clz.academy = academy
        print("í´ë˜ìŠ¤ ë©”ì†Œë“œì—ì„œ {}ë¡œ í´ë˜ìŠ¤ ë³€ìˆ˜ ë³€ê²½".format(clz.academy))
    @staticmethod
    def that_():  # í´ë˜ìŠ¤ ë³€ìˆ˜ë³´ë‹¤ëŠ” ë¡œì§ì´ ìš°ì„ ì¼ ê²½ìš°
        Person.academy = "staticAcademy"
        print("static ë©”ì†Œë“œì—ì„œ {}ë¡œ í´ë˜ìŠ¤ ë³€ìˆ˜ ë³€ê²½".format(Person.academy))
```
